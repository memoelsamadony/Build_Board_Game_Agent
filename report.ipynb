```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
```

```python
def parse_txt(txt):
    txt = re.sub(r'\[|\]|,', '', txt)
    return float(txt)
```

```python
# Read the data from the file
with open("data.txt", "r") as f:
    player_one_times = f.readline().split()
    player_two_times = f.readline().split()
    player_one_nodes = f.readline().split()
    player_two_nodes = f.readline().split()


player_one_times = [parse_txt(x) for x in player_one_times if x != ',']
player_two_times = [parse_txt(x) for x in player_two_times if x != ',']
player_one_nodes = [parse_txt(x) for x in player_one_nodes if x != ',']
player_two_nodes = [parse_txt(x) for x in player_two_nodes if x != ',']


```

```python
# Create a figure and an axes object
fig, ax = plt.subplots()

# Plot the two arrays of times on the axes object
ax.plot(player_one_times, label="Player One")
ax.plot(player_two_times, label="Player Two")
ax.legend()
# Set the title of the plot
ax.set_title("Time Taken to Make a Move")

# Set the x-axis label
ax.set_xlabel("Move Number")

# Set the y-axis label
ax.set_ylabel("Time (Seconds)")

# ax.yaxis.set_ticks(np.arange(0, max(player_one_times + player_two_times), 10))

# # Set the y-axis labels to be at every 10th value
# ax.yaxis.set_ticklabels(np.arange(0, max(player_one_times + player_two_times), 10))

# Apply a beautiful theme from seaborn
sns.set_theme(style="darkgrid")
plt.show();
```

```python
fig, ax = plt.subplots()

# Plot the two arrays of times on the axes object
ax.plot(player_one_nodes, label="Player One")
ax.plot(player_two_nodes, label="Player Two")
ax.legend()
# Set the title of the plot
ax.set_title("Number of Nodes Expanded vs Move Number")

# Set the x-axis label
ax.set_xlabel("Move Number")

# Set the y-axis label
ax.set_ylabel("Number of Nodes Expanded")

# ax.yaxis.set_ticks(np.arange(0, max(player_one_times + player_two_times), 10))

# # Set the y-axis labels to be at every 10th value
# ax.yaxis.set_ticklabels(np.arange(0, max(player_one_times + player_two_times), 10))

# Apply a beautiful theme from seaborn
sns.set_theme(style="darkgrid")
plt.show();
```

```python
fig, ax = plt.subplots()

# Calculate the average of each variable
player_one_nodes_avg = np.mean(player_one_nodes)
player_two_nodes_avg = np.mean(player_two_nodes)

# Plot the two arrays of averages on the axes object
ax.bar(["Player One", "Player Two"], [player_one_nodes_avg, player_two_nodes_avg], width=0.5)
ax.set_title("Average Number of Nodes Expanded in a Game")

# Set the x-axis label
ax.set_xlabel("players")

# Set the y-axis label
ax.set_ylabel("Average Number of Nodes Expanded")

# Apply a beautiful theme from seaborn
sns.set_theme(style="darkgrid")
plt.show();
```

```python
fig, ax = plt.subplots()

# Calculate the average of each variable
player_one_time_avg = np.mean(player_one_times)
player_two_time_avg = np.mean(player_two_times)

# Plot the two arrays of averages on the axes object
ax.bar(["Player One", "Player Two"], [player_one_time_avg, player_two_time_avg], width=0.5)
ax.set_title("Average Time Taken to generate a move throughout the Game")

# Set the x-axis label
ax.set_xlabel("players")

# Set the y-axis label
ax.set_ylabel("Average Time Taken to generate a move")

# Apply a beautiful theme from seaborn
sns.set_theme(style="darkgrid")
plt.show();
```
